```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
!rm -r /content/spark-warehouse/langchain_spark.db
```

```python
schema = 'langchain_spark'
tbl_nm = 'stroke_data'
spark.sql(f"CREATE DATABASE IF NOT EXISTS {schema}")
df = spark.read.csv('/content/healthcare-dataset-stroke-data.csv', header=True, inferSchema=True)
df.write.saveAsTable(f"{schema}.{tbl_nm}")
```

```python
spark.table(f"{schema}.{tbl_nm}").show()
```

```python
from langchain.agents import create_spark_sql_agent
from langchain.agents.agent_toolkits import SparkSQLToolkit
from langchain.utilities import SparkSQL
```

```python
from langchain.llms import LlamaCpp
```

```python
from huggingface_hub import hf_hub_download
#hf_hub_download(repo_id='TheBloke/Llama-2-7B-GGUF', filename='llama-2-7b.Q2_K.gguf')
hf_hub_download(repo_id='TheBloke/Llama-2-7b-Chat-GGUF', filename='llama-2-7b-chat.Q2_K.gguf')
```

```python
llm = LlamaCpp(
    #model_path = '/root/.cache/huggingface/hub/models--TheBloke--Llama-2-7B-GGUF/snapshots/61289dfe7c8355b61c7ddd6178e89300999b11eb/llama-2-7b.Q2_K.gguf',
    model_path='/root/.cache/huggingface/hub/models--TheBloke--Llama-2-7b-Chat-GGUF/snapshots/9ca625120374ddaae21f067cb006517d14dc91a6/llama-2-7b-chat.Q2_K.gguf',
    n_gpu_layers=40,
    n_batch=512,
    n_ctx=1024,
    verbose=True
)
```

```python
spark_sql = SparkSQL(schema=schema)
toolkit = SparkSQLToolkit(db=spark_sql, llm=llm)
agent_executor = create_spark_sql_agent(llm=llm, toolkit=toolkit, verbose=True)
```

```python
agent_executor.run("Describe the stroke_data table")
```

```python
agent_executor.run("How many persons have heart disease?")
```
